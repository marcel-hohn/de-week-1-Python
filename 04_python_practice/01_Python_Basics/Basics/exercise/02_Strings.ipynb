````python
# 1 middle chars
# a function that returns the middle char if length is odd and two middle chars if length is even
def middle(s):
    n = len(s)
    if n % 2 == 1:
        return s[n//2]
    else:
        return s[n//2 - 1: n//2 + 1]
    
assert middle("abc") == "b"
assert middle("abba") == "bb"
````

````python
# 2 Title-case safely**
# `safe_title(s)` → title case but do not lowercase words fully in ALL CAPS acronyms (e.g., `"learn SQL now"` → `"Learn SQL Now"`; `"use GPU"` → `"Use GPU"`).
def safe_title(s):
    words = s.split()
    for i in range(len(words)):
        if words[i].isupper():
            continue
        words[i] = words[i].capitalize()
    return " ".join(words)

safe_title('learn SQL now')
````

````python
# 3 **Reverse words**
#* `reverse_words(s)` → reverse word order but keep internal characters.
def reverse_words(s):
    if s == "":
        return s
    parts = []
    i = 0
    n = len(s)
    while i < n:
        j = i
        if s[i].isspace():
            while j < n and s[j].isspace():
                j += 1
        else:
            while j < n and not s[j].isspace():
                j += 1
        parts.append(s[i:j])
        i = j
    # tausche die Wortteile (die Segmente, die nicht mit Whitespace beginnen)
    word_indices = [k for k, seg in enumerate(parts) if not seg[0].isspace()]
    words = [parts[k] for k in word_indices][::-1]
    for idx, k in enumerate(word_indices):
        parts[k] = words[idx]
    return ''.join(parts)

assert reverse_words("one two  three") == "three two  one"
````

````python
# 4. **Normalize spaces**
# 
def squash_spaces(s):
   return " ".join(s.split())

assert squash_spaces("  a\t b   c ") == "a b c"
````

````python
# 5. **CSV line → list (no quotes)**
# * `split_csv(s)` → split by commas, trim whitespace around fields; ignore empty trailing field if line ends with comma.
def split_csv(s):
    result = []
    for field in s.split(','):
        if field.strip():
            result.append(field.strip())
    return result
assert split_csv(" a, b ,c,") == ["a","b","c"]
````

````python
# 6 **Mask secrets**
def mask_email(s):
    user, domain = s.split('@')
    domain_parts = domain.split('.')
    tld = domain_parts[-1]
    masked_user = user[0] + '*' * (len(user) - 1)
    masked_domain = domain_parts[0][0] + '*' * (len(domain_parts[0]) - 1) + '.' + tld
    return f"{masked_user}@{masked_domain}"
mask_email("alice@example.com")
````

````python
# 7 **Find all indexes**
# * `find_all(s, sub)` → list of start indices where `sub` occurs (including overlaps).

def find_all(s, sub):
    result = []
    for i in range(len(s) - len(sub) + 1):
        if s[i:i+len(sub)] == sub:
            result.append(i)
    return result

assert find_all("aaaa", "aa") == [0,1,2]
````

````python
#8. **Anagram check**
def is_anagram(a, b):
    # Remove spaces and convert to lowercase
    a = a.replace(" ", "").lower()
    b = b.replace(" ", "").lower()

    # Check if lengths are equal
    if len(a) != len(b):
        return False

    # Sort characters and compare
    return sorted(a) == sorted(b)

assert is_anagram("Listen", "Silent")
````

````python
#9 **Format table row**
# * `fmt_row(values, widths)` → left-align strings to fixed widths, joined by `" | "`.
def fmt_row(values, widths):
    result=[]
    for i in range(len(values)):
        result.append(values[i] + (widths[i] - len(values[i])) * " ")
    return " | ".join(result)
assert fmt_row(["a","bb"], [3,4]) == "a   | bb  "
````

````python
#10 Challenge **Slugify**
# * `slugify(title)` → lowercase, trim, replace runs of non-alnum with single `-`, remove leading/trailing `-`.
def slugify(title):
    title = title.lower().strip()

    result = ""
    prev_dash = False

    for c in title:
        if c.isalnum():
            result += c
            prev_dash = False
        else:
            if not prev_dash:
                result += "-"
                prev_dash = True

    result = result.strip("-")
    return result

assert slugify("Hello,  World!!") == "hello-world"
````
