````python
# 1
def num_type(x):
    if type(x) == int:
        return "int"
    if type(x) == float:
        return "float"
    if type(x) == bool:
        return "bool"

assert num_type(3) == "int"
assert num_type(2.0) == "float"
assert num_type(True) == "bool"
````

````python
# 2 Safe division
def safe_div(a, b, default=None):
    if b==0:
        return default
    else:
        return a/b
assert safe_div(6, 3) == 2.0
assert safe_div(1, 0, default=float('inf')) == float('inf')
````

````python
# 3 Integer vs float division
def divisions(a, b):
    return (a // b, a / b, a % b)
assert divisions(7, 3) == (2, 7/3, 1)
assert divisions(-7, 3)[0] == -3   # floor division
````

````python
# 4 Rounding modes
def bankers_round(x, ndigits=0):
    return round(x,ndigits)

def classic_round(x, ndigits=0):
    factor = 10 ** ndigits
    x = x * factor
    
    if x>=0:
        result = int(x + 0.5)
    else:
        result = int(x-0.5)

    return result/factor

assert bankers_round(2.5) == 2
assert classic_round(2.5) == 3
````

````python
# 5 Parse numbers

def parse_int(s, base=10, default=None):
    try:
        return int(s.strip(),base)
    except ValueError:
        return default
    
assert parse_int(" 1_024 ") == 1024
assert parse_int("FF", base=16) == 255
assert parse_int("oops", default=-1) == -1
````

````python
# 6 Boolean arithmetic

def true_ratio(seq):
    x=0 # start for true counter
    for i in seq: # for loop to loop over sequence
        if bool(i)==True: # bool() converts seq value to boolian, if checks for True
            x+=1
    return x/len(seq) # devide by seq length to get ratio
    
assert abs(true_ratio([True, False, 1, 0, "", "x"]) - 0.5) < 1e-9
````

````python
# 7 Min/Max & aggregates
# Write stats(nums) → dict with count, total, mean, minimum, maximum. Empty list → return zeros/None appropriately.

def stats(nums):
    return {
        "count": len(nums),
        "total": sum(nums),
        "mean": sum(nums)/len(nums),
        "min": min(nums),
        "max": max(nums),
    }
    ...
assert stats([1,2,3])["mean"] == 2
````

````python
# 8 Bucket by step
# Write bucket(x, step) → largest multiple of step ≤ x (works with negatives).

def bucket(x, step):
    return (x//step) * step

assert bucket(37, 10) == 30
assert bucket(-3, 5) == -5
````

````python
# 9 Time math (seconds → h:m:s)
# Write to_hms(seconds) → "HH:MM:SS" with zero-padding.

def to_hms(seconds):
    HH = seconds//3600
    MM = (seconds%3600) // 60
    SS = seconds % 60
    return f"{HH:02}:{MM:02}:{SS:02}"
    
    ...
assert to_hms(3661) == "01:01:01"
````

````python
# 10 Running totals with precise rounding
# Write running_totals(amounts, ndigits=2) that returns cumulative sums rounded at each step (like financial ledgers) using banker’s rounding. 
#Example: [10.005, 0.005] → [10.01, 10.02].

def running_totals(amounts, ndigits=2):
    x = 0
    result = []
    
    for i in amounts:
        x += i
        x = round(x, ndigits)
        result.append(x)
        
    return result
assert running_totals([10.005, 0.005], 2) == [10.01, 10.02]
````
